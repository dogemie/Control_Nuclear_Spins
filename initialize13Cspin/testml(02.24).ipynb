```python
import torch
import torch.nn.functional as F
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('C:/Users/Administrator/Dogyeom(2023.01.01)/KIST_intern/Task1/Control_Nuclear_Spins/initialize13Cspin/trashdata/testdata_second.csv')
df
```

```python
y = df[['x', 'N', 'z']]
# y

# y = df[['Xtau', 'XN', 'Ztau', 'ZN']]
# y
```

```python
X = df[['Al', 'Ap']]
# X
```

```python
Xdata = list(np.array(X.values.tolist()))
# print(Xdata)
```

```python
ydata = list(np.array(y.values.tolist()))
# print(ydata)
```

```python
# Dataset 상속
class CustomDataset(Dataset): 
  def __init__(self):
    self.x_data = Xdata
    self.y_data = ydata

  # 총 데이터의 개수를 리턴
  def __len__(self): 
    return len(self.x_data)

  # 인덱스를 입력받아 그에 맵핑되는 입출력 데이터를 파이토치의 Tensor 형태로 리턴
  def __getitem__(self, idx): 
    x = torch.FloatTensor(self.x_data[idx])
    y = torch.FloatTensor(self.y_data[idx])
    return x, y
```

```python
dataset = CustomDataset()
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
model = torch.nn.Linear(2,3)
optimizer = torch.optim.SGD(model.parameters(), lr=1e-5) 
```

```python
nb_epochs = 1000
for epoch in range(nb_epochs + 1):
  for batch_idx, samples in enumerate(dataloader):
    # print(batch_idx)
    # print(samples)
    x_train, y_train = samples
    # H(x) 계산
    prediction = model(x_train)

    # cost 계산
    cost = F.mse_loss(prediction, y_train)

    # cost로 H(x) 계산
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print('Epoch {:4d}/{} Batch {}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, batch_idx+1, len(dataloader),
        cost.item()
        ))
```

```python
# 임의의 입력 [73, 80, 75]를 선언
new_var =  torch.FloatTensor([[4.493041, 0.5427]]) 
# 입력한 값 [73, 80, 75]에 대해서 예측값 y를 리턴받아서 pred_y에 저장
pred_y = model(new_var) 
print("훈련 후 입력이 73, 80, 75일 때의 예측값 :", pred_y) 
```
