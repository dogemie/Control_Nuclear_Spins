```python
import numpy as np
from qutip import Qobj
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from qutip.operators import sigmax, sigmay, sigmaz

def fidelity(rho1, rho2):
    """Calculates the fidelity between two density matrices"""
    sqrt_rho1 = Qobj(rho1) ** 0.5
    return (sqrt_rho1 * Qobj(rho2) * sqrt_rho1) ** 0.5

def plot_fidelity_on_bloch(rho1, rho2):
    """Plots the fidelity between two density matrices on the Bloch sphere"""
    # Define the Bloch vectors
    bloch_vectors = [np.array([np.real(np.trace(Qobj(rho).full() @ sigmax)),
                               np.real(np.trace(Qobj(rho).full() @ sigmay)),
                               np.real(np.trace(Qobj(rho).full() @ sigmaz))])
                     for rho in [Qobj(rho1), Qobj(rho2)]]
    
    colors = ['red', 'blue']
    for vec, color in zip(bloch_vectors, colors):
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        ax.plot([0, vec[0]], [0, vec[1]], [0, vec[2]], color=color, linewidth=3)
        ax.set_xlim([-1, 1])
        ax.set_ylim([-1, 1])
        ax.set_zlim([-1, 1])
        ax.set_aspect('equal')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        ax.set_title('Fidelity on Bloch sphere')
        plt.show()
        
# Example usage
rho1 = np.array([[0.5, 0.5], [0.5, 0.5]])
rho2 = np.array([[1, 0], [0, 0]])

# Plot fidelity on bloch sphere
plot_fidelity_on_bloch(rho1, rho2)
```
